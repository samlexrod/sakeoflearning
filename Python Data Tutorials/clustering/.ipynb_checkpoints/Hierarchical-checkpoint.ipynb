# Import Always First

```python
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy as hpl
from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
import seaborn as sns
```

# Extracting Sci-kit Learn Data

```python
iris = datasets.load_iris()
iris.data[:10]
```

```python
iris.target
```

```python
iris.target_names
```

# Clustering Iris Data

```python
model = AgglomerativeClustering(n_clusters=3)
pred = model.fit_predict(iris.data)
```

```python
adjusted_rand_score(iris.target, pred)
```

# Clustering Normalized Data

```python
iris.norm_data = normalize(iris.data)
iris.norm_data[:10]
```

```python
model_norm = AgglomerativeClustering(n_clusters=3)
norm_pred = model_norm.fit_predict(iris.norm_data)
```

```python
adjusted_rand_score(iris.target, norm_pred)
```

# Assign Another Type of Linkage Other than the Default Ward

```python
model_norm_comp = AgglomerativeClustering(n_clusters=3, linkage='complete')
norm_pred_comp = model_norm_comp.fit_predict(iris.norm_data)
```

```python
adjusted_rand_score(iris.target, norm_pred_comp)
```

# Plotting a Dendogram Using Scipy

```python
linkage_matrix = hpl.linkage(iris.norm_data, 'ward')
plt.figure(figsize=(22, 18))
hpl.dendrogram(linkage_matrix)
plt.show()
```

# Plotting Clustermap Using Seaborn

```python
sns.clustermap(iris.norm_data, figsize=(12, 18), method='ward', cmap='viridis')
plt.show()
```
