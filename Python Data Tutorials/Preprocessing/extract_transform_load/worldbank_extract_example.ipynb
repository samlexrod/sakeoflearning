# Imports Always Frist

```python
import requests
import pandas as pd
import json
```

# Extracting Data from Wold Bank

```python
# Setting indicators and countries
indicator_code = 'SP.POP.TOTL'
country_abbr = 'sz;us'

# The url
url = f'http://api.worldbank.org/v2/countries/{country_abbr}/indicators/{indicator_code}/?format=json&per_page=1000'

# Start request
r = requests.get(url)
```

# Loading Json Data into Dataframe

```python
pd.DataFrame(r.json()[1]).where(
    lambda x: x!='').assign(
    country=lambda x: x.country.apply(lambda x: x.get('value')),
    indicator=lambda x: x.indicator.apply(lambda x: x.get('value'))).drop(
    ['decimal'], axis=1).dropna(
    how='all', axis=1)
```
