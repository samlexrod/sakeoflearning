**Purpose**: to group abstract objects into classes of similar objects 

# Imports Always First

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score
```

# Simulate Cluster Data

```python
make_blobs?
```

```python
dataset, target = make_blobs(n_samples=1000,
                            cluster_std=[5, 1, 0.5],
                            random_state=42)
```

```python
dataset
```

# Visualizing Actual Data

```python
plt.scatter?
```

```python
plt.title?
```

```python
plt.figure(figsize=(16, 12))
plt.scatter(dataset[:, 0], dataset[:, 1],
           c=target,
           edgecolor='grey',
           linewidths =1.5,
           s=100,
           cmap=plt.get_cmap('viridis'))
plt.title('Actual Data', fontdict={'fontsize': 30})
plt.show()
```

# Creating Kmeans

```python
model_km = KMeans(n_clusters=3)
pred_km = model_km.fit_predict(dataset)
```

# Visualizing Predicted Data

```python
plt.figure(figsize=(16, 12))
plt.scatter(dataset[:, 0], dataset[:, 1],
           c=pred_km,
           edgecolor='grey',
           linewidths =1.5,
           s=100,
           cmap=plt.get_cmap('viridis'))
plt.title('Predicted Data', fontdict={'fontsize': 30})
plt.show()
```

```python
adjusted_rand_score(target, pred_km)
```
