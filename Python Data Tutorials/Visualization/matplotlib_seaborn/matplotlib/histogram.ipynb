# Imports Always First

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Creating a Dataframe

```python
two_normal = np.random.randn(1000, 2)
two_normal[:20]
```

```python
one_random = np.random.rand(1000, 1)
one_random[:20]
```

```python
df = pd.concat([pd.DataFrame(two_normal), pd.DataFrame(one_random)], axis=1)
df.columns = ['Normal1', 'Normal2', 'Random']
df.head()
```

# Creating the Histograms

## Using Pandas

```python
df.hist(figsize=(16, 10), grid=False)
plt.show()
```

## Using Matplotlib

```python
plt.figure(figsize=(16, 10))
plt.hist(df.Normal1)
plt.title(df.Normal1.name)
plt.show()
```
