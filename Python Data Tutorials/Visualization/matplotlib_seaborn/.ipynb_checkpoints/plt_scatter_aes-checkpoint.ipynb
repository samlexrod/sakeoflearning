```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from IPython.display import IFrame
```

# Simulating Cluster Data

```python
dataset, target = make_blobs(n_samples=1000, cluster_std=[5, 1, 0.5], random_state=42)
```

```python
dataset
```

```python
target
```

# Plotting Data on Scatterplot

```python
?plt.scatter
```

```python
?plt.get_cmap
```

```python
plt.figure(figsize=(16, 12))
plt.scatter(dataset[:, 0], dataset[:, 1], 
            c=target, # for colors
            edgecolors='grey', # for edges
            linewidths=1.5, # for line width of edges
            s=100, # for shape size
            marker='o',
            cmap=plt.get_cmap('viridis')) # for color style
plt.show()
```
