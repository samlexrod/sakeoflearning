# Imports Always Frist

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Creating Data

```python
df = pd.DataFrame(np.random.randn(10, 5))
```

# Creating Heatmap

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True, fmt=".3f", vmin=-1, vmax=1);
```
