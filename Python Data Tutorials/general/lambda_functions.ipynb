A lambda function is a function that is not defined.

Take for example a defined function:
```python
def defined_function(parameter1, parameter2):
    print(parameter1 * parameter2)
```

The same thing can be accomplished by using a lambda function:
```python
lambda parameter1, parameter2: print(parameter1, parameter2)
```

````python
def defined_function(parameter1, parameter2):
    print(parameter1 * parameter2)
defined_function(1, 2)
````

````python
undefined_function = lambda parameter1, parameter2: print(parameter1 * parameter2)
defined_function(1, 2)
````
